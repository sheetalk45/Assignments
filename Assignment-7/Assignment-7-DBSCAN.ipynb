# EastWestAirlines

```python
#Import the libraries
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import .csv file and convert it to a DataFrame object
df = pd.read_csv("EastWestAirlines.csv");

print(df.head())
```

```python
print(df.info())
```

```python
df.drop(['ID#'],axis=1,inplace=True)
```

```python
df
```

```python
stdscaler = StandardScaler().fit_transform(df)
stdscaler
```

```python
# DBSCAN Clustering
dbscan=DBSCAN(eps=1,min_samples=6)
dbscan.fit(df)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

```python
# Making clusters column (optional step)
clusters=pd.DataFrame(dbscan.labels_,columns=['clusters'])
clusters
```

```python
# Adding clusters to dataset
df['Clusters']=dbscan.labels_
df
```

# Crime_data

```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
# Import Dataset
crime=pd.read_csv('crime_data.csv')
crime
```

```python
crime.info()
```

```python
crime.drop(['Unnamed: 0'],axis=1,inplace=True)
crime
```

```python
# Normalize heterogenous numerical data using standard scalar fit transform to dataset
crime_norm=StandardScaler().fit_transform(crime)
crime_norm
```

```python
# DBSCAN Clustering
dbscan=DBSCAN(eps=1,min_samples=4)
dbscan.fit(crime_norm)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

```python
# Adding clusters to dataset
crime['clusters']=dbscan.labels_
crime
```

```python
crime.groupby('clusters').agg(['mean']).reset_index()
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))  
plt.scatter(crime['clusters'],crime['UrbanPop'], c=dbscan.labels_) 
```
